```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

# Parameter
n = 100
phi = 0.2
sigma = 1.0

# Fungsi simulasi AR(1)
def simulate_ar1(phi, n, sigma=1.0):
    """
    Simulasi AR(1):
    Y_t = mu + phi * Y_{t-1} + e_t
    e_t ~ N(0, sigma^2)
    """
    Y = np.zeros(n)
    et = np.random.normal(0, sigma, n)

    for t in range(1, n):
        Y[t] = phi * Y[t-1] + et[t]

    return Y

# Jalankan simulasi
Y = simulate_ar1(phi, n, sigma)

# Plot hasil simulasi
plt.figure()
plt.plot(Y)
plt.title("Simulasi AR(1) dengan φ = 0.2")
plt.xlabel("Waktu")
plt.ylabel("Y_t")
plt.grid()
plt.show()

model = AutoReg(Y, lags=1)
model_fit = model.fit()

print(model_fit.summary())
```
